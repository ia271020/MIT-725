**QUESTION 2: Use appropriate data mining methodologies and clean the dataset downloaded**

```python
df = df.drop("Name", axis=1)

df.head()
```

The column Name was removed from the dataset because it represents country names, which do not contribute to predicting transmission classification. Removing irrelevant attributes improves model performance and reduces unnecessary complexity.

```python
df.isnull().sum()
```

```python
df = df.dropna()

df.isnull().sum()
```

Missing values were identified in several variables within the dataset. These missing observations were removed using the dropna() function to ensure compatibility with machine learning classifiers such as K-Nearest Neighbour and Support Vector Machine, which do not support missing values directly. This step improved the reliability of the dataset before model training.

```python
from sklearn.preprocessing import LabelEncoder

le_region = LabelEncoder()
le_target = LabelEncoder()

df["WHO Region"] = le_region.fit_transform(df["WHO Region"])
df["Transmission Classification"] = le_target.fit_transform(df["Transmission Classification"])

df.head()
```

Categorical variables such as WHO Region and Transmission Classification were converted into numerical format using Label Encoding. Machine learning algorithms require numerical input values for training classification models, therefore encoding categorical variables improves compatibility with classification techniques.

```python
X = df.drop("Transmission Classification", axis=1)
y = df["Transmission Classification"]

print("Features shape:", X.shape)
print("Target shape:", y.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

X_scaled[:5]
```

Feature scaling was performed using the StandardScaler technique to standardize the dataset variables. This step ensures that all features contribute equally during model training, especially for distance-based classifiers such as K-Nearest Neighbour and Support Vector Machine.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.3,
    random_state=42
)

print("Training set size:", X_train.shape)
print("Testing set size:", X_test.shape)
```

The dataset was divided into training and testing subsets using the train_test_split function. Seventy percent (70%) of the dataset was used for training the models, while thirty percent (30%) was reserved for testing model performance. This ensures reliable evaluation of classifier accuracy.

QUESTION 3: Use Decision Trees, Random Forest, K-Nearest Neighbour and Support Vector Machine classifiers to predict different Transmission Classification of the pandemic by determining the performance metrics such as precision, recall, F1-score, accuracy and Jaccard index.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score, jaccard_score

dt_model = DecisionTreeClassifier(random_state=42)

dt_model.fit(X_train, y_train)

dt_predictions = dt_model.predict(X_test)

print(classification_report(y_test, dt_predictions))
print("Accuracy:", accuracy_score(y_test, dt_predictions))
print("Jaccard Index:", jaccard_score(y_test, dt_predictions, average='weighted'))
```

The Decision Tree classifier was applied to predict transmission classification using the cleaned dataset. The model performance was evaluated using precision, recall, F1-score, accuracy and Jaccard similarity index. The results show that the Decision Tree classifier achieved moderate classification performance, demonstrating its ability to identify transmission classification patterns within the dataset. However, Decision Trees may experience some misclassification due to their sensitivity to dataset variations compared to ensemble learning techniques.

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, dt_predictions)
```

The confusion matrix illustrates the classification performance of the Decision Tree model by comparing predicted transmission classifications with the actual values. The diagonal elements represent correctly classified observations, while off-diagonal values indicate misclassifications between transmission categories.

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(random_state=42)

rf_model.fit(X_train, y_train)

rf_predictions = rf_model.predict(X_test)

print(classification_report(y_test, rf_predictions))
print("Accuracy:", accuracy_score(y_test, rf_predictions))
print("Jaccard Index:", jaccard_score(y_test, rf_predictions, average='weighted'))
```

The Random Forest classifier was applied to improve prediction accuracy by combining multiple decision trees through an ensemble learning approach. The performance metrics obtained, including precision, recall, F1-score, accuracy and Jaccard similarity index, indicate improved classification performance compared to the Decision Tree model. Random Forest reduces overfitting and increases prediction reliability by aggregating results from multiple trees.

```python
confusion_matrix(y_test, rf_predictions)
```

The confusion matrix for the Random Forest classifier shows the number of correct and incorrect predictions made by the model across the different transmission classification categories. Higher values along the diagonal indicate stronger classification accuracy compared to the Decision Tree model.

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors=5)

knn_model.fit(X_train, y_train)

knn_predictions = knn_model.predict(X_test)

print(classification_report(y_test, knn_predictions))
print("Accuracy:", accuracy_score(y_test, knn_predictions))
print("Jaccard Index:", jaccard_score(y_test, knn_predictions, average='weighted'))
```

The K-Nearest Neighbour classifier was applied to predict transmission classification using distance-based similarity between observations. The model performance was evaluated using precision, recall, F1-score, accuracy and Jaccard similarity index. The results indicate moderate classification performance. Since KNN relies on similarity between neighbouring data points, its performance depends on dataset structure and feature scaling.

```python
confusion_matrix(y_test, knn_predictions)
```

The confusion matrix for the K-Nearest Neighbour classifier shows the distribution of correct and incorrect predictions across the transmission classification categories. The diagonal elements represent correctly classified observations, while off-diagonal elements indicate classification errors between categories.

```python
from sklearn.svm import SVC

svm_model = SVC()

svm_model.fit(X_train, y_train)

svm_predictions = svm_model.predict(X_test)

print(classification_report(y_test, svm_predictions))
print("Accuracy:", accuracy_score(y_test, svm_predictions))
print("Jaccard Index:", jaccard_score(y_test, svm_predictions, average='weighted'))
```

The Support Vector Machine classifier was applied to predict transmission classification by identifying optimal decision boundaries between classes. The model performance was evaluated using precision, recall, F1-score, accuracy and Jaccard similarity index. The results indicate moderate classification performance. Support Vector Machine classifiers perform well when classes are clearly separable, but performance may reduce when there is overlap between transmission classification categories.

```python
confusion_matrix(y_test, svm_predictions)
```

The confusion matrix for the Support Vector Machine classifier shows the number of correctly and incorrectly predicted transmission classification categories. The diagonal values represent correct classifications, while off-diagonal values indicate misclassification between categories.
